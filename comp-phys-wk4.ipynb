```python
y = np.linspace(1e-6, 15, 1000)
b = y**3/(np.exp(y)-1)
plt.plot(y, b)
plt.xlabel('Frequency')
plt.ylabel('Energy Density')
plt.title('Black Body Spectrum')
plt.savefig('planck.pdf')
plt.show()
```

```python
y = np.linspace(1e-6, np.pi, 1000)
b = np.cos(-np.exp(y))
plt.plot(y, b)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Frequency Modulation')
plt.savefig('modulation.pdf')
plt.show()
```

```python
from pylab import *
Xmin = -5.0; Xmax = 5.0
Npoints = 500
DelX = (Xmax-Xmin)/Npoints
x = np.arange(Xmin, Xmax, DelX)
y1 = -np.sin(x)*np.cos(x**2)
y2 = np.exp(-x/4.)*np.sin(x)

fig = plt.figure()
ax = fig.add_subplot(211)
fig.subplots_adjust(top=0.85)
plt.title('Two Functions')

plt.plot(x, y1, 'r.', lw=2)
plt.grid(True)
ax.set_ylabel('y1')

plt.subplot(212)
plt.plot(x, y2, 'g--', lw=2)
plt.xlabel('x')
plt.ylabel('y2')

plt.show()
```

```python
def g(x, y):
    return x**2+y**2
z = g(2., 3.)
print z

%matplotlib osx
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.linspace(-5, 5, 100)

xv, yv = np.meshgrid(x, y)
z = g(xv, yv)
#ax.plot_wireframe(xv, yv, z, rstride = 6, cstride = 6, linewidth = 1)
ax.plot_surface(xv, yv, z, rstride = 2, cstride = 2, cmap=cm.coolwarm, linewidth = 0)
plt.show()
```
